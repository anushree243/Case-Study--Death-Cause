#### Top states in which Deaths caused by cancer

```python
df.head()
```

```python
df['Cause Name'].unique()
```

```python
data = df[(df['Cause Name']=='Cancer') & (df['State']!='United States')]
```

```python
data.head()
```

```python
data['Cause Name'].unique()
```

```python
data2 = data.groupby(by=['State']).sum()
```

```python
data2.head()
```

```python
data2.reset_index(inplace=True)
```

```python
data2.sort_values(by='Deaths',ascending=False, inplace=True)
```

```python
data2.head()
```

```python
fig,ax = plt.subplots(figsize=(7,6))
ax.bar('State','Deaths', data=data2.head(10))
ax.set_xlabel('State', fontsize=20)
ax.set_ylabel('Deaths',fontsize=20)
ax.set_title('Top 10 States where death occured because of Cancer',fontsize=24, color='purple')
```

```python
# Top 10 States according to Deaths 
plot2 = df[df['State']!='United States']
plot2 = plot2.groupby(by=['State'])['Deaths'].sum().reset_index().sort_values(by='Deaths',ascending=False)
plot2
```

```python
plot1 = df.groupby('Year')['Deaths'].sum().reset_index()
plot1
```

```python
# Top 3 Causes of Death
plot3 = df[df['Cause Name']!='All Causes']
plot3 = plot3.groupby('Cause Name')['Deaths'].sum().reset_index().sort_values('Deaths',ascending=False)
plot3
```

```python
fig, ax = plt.subplots(2,2, figsize=(20,6))
ax[0][0].bar('Year','Deaths', data=plot1.head(10))
ax[0][1].bar('State','Deaths', data=plot2.head(10))
ax[1][0].bar('Cause Name','Deaths', data=plot3.head(3))
ax[1][1].bar('State','Deaths', data=plot2.tail(10))
```

```python
plot1=plot1
plot2,plot4=plot2.head(10),plot2.tail(10)
plot3=plot3.head(3)
fig, ax = plt.subplots(2,2, figsize=(23,10))
ax[0][0].bar(plot1['Year'],plot1['Deaths'], color='purple')
ax[0,0].set_xlabel('Year',fontsize=20)
ax[0,0].set_ylabel('Deaths',fontsize=20)
ax[0,0].set_title('Deaths Each Year from 1999-2017',fontsize=20)

ax[0][1].bar(plot2['State'],plot2['Deaths'], color='darkblue')
ax[0,1].set_xlabel('State',fontsize=15)
ax[0,1].set_ylabel('Deaths',fontsize=15)
ax[0,1].set_title('Top 10 State acc to Deaths',fontsize=20)
plt.xticks(rotation=45)

ax[1][0].bar(plot3['Cause Name'],plot3['Deaths'],color='c')
ax[1,0].set_xlabel('Cause',fontsize=15)
ax[1,0].set_ylabel('Deaths',fontsize=15)
ax[1,0].set_title('Top 3 Cause of Death',fontsize=20)

ax[1][1].bar(plot4['State'],plot4['Deaths'],color='r')
ax[1,1].set_xlabel('State',fontsize=15)
ax[1,1].set_ylabel('Deaths',fontsize=15)
ax[1,1].set_title('Bottom 10 State acc to Deaths',fontsize=20)

plt.xticks(rotation=45)
plt.tight_layout()
```
